```python
import numpy as np
import healpy as hp
import pandas as pd
import matplotlib.pyplot as plt
import tools
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
bin_ellmin, bin_ellmax, bin_width = 20, 7020, 20
ells = np.arange(bin_ellmin, bin_ellmax+1, 1)
smooth_ellmin, smooth_ellmax, smooth_width = 30, 7000, 1
smooth_ells = np.arange(smooth_ellmin, smooth_ellmax+1, smooth_width)
```

```python
lensed_cmb_file = './maps/Colin/original/Sehgalsimparams_healpix_4096_KappaeffLSStoCMBfullsky_phi_SimLens_Tsynfastnopell_fast_lmax8000_nside4096_interp2.5_method1_1_lensed_map.fits'
phi_file = './maps/Colin/original/healpix_4096_KappaeffLSStoCMBfullsky_phi.fits'
ksz_file = './maps/Colin/original/148_ksz_healpix_nopell_Nside4096_DeltaT_uK.fits'
```

```python
# write lensed_cmb_alms_7000.fits
lensed_cmb = hp.read_map(lensed_cmb_file)
lensed_cmb_alms = hp.map2alm(lensed_cmb, lmax=smooth_ellmax)
hp.write_alm('maps/Colin/alms/lensed_cmb_alms_%s.fits' %(smooth_ellmax), lensed_cmb_alms, \
             lmax=smooth_ellmax, overwrite=True)
```

```python
ksz_lt_cls = hp.alm2cl(ksz_lt_alms)
```

```python
Smooth_ksz = tools.bin_smooth(bin_ells, ksz_lt_cls[bin_ells], bin_width)
smooth_ksz_lt_cls = Smooth_ksz.smooth(smooth_ells)
```
